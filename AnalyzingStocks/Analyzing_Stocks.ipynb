```python
df_stocks["Part. (%)"] = df_stocks["Part. (%)"].apply(lambda x: float(x.replace(",",".")))
```

```python
stocks = df_stocks.sort_values("Part. (%)").tail(10)["Código"].values
stocks
```

```python
tickers = yf.Tickers(" ".join([i + '.SA' for i in stocks]))
tickers.tickers
```

```python
dict_stocks = {}

for ticker in tickers.tickers.keys():
     dict_stocks[ticker]= tickers.tickers[ticker].history(period="max")
```

```python
# Combining all data in an unique DataFrame
```

```python
df_data = pd.DataFrame()
```

```python
for ticker in dict_stocks.keys():
    df_aux = dict_stocks[ticker].rename(columns={"Close":ticker})
    df_data = df_data.join(df_aux[ticker],how="outer")
```

```python
df_check = df_return[df_return.shift(1) < 0]

fig, ax = plt.subplots(figsize=(20,8))
sns.histplot(df_check, kde=True)

print("Mean of returns")
for stock in stocks:
    print('{} - {}'.format(stock, df_check[stock + '.SA'].mean()*100))

print("\nStandard deviation of returns")
for stock in stocks:
    print('{} - {}'.format(stock, df_check[stock + '.SA'].std()*100))
```

```python
df_check = df_return[df_return.shift(1) < -0.05]

fig, ax = plt.subplots(figsize=(20,8))
sns.histplot(df_check, kde=True)

print("Mean of returns")
for stock in stocks:
    print('{} - {}'.format(stock, df_check[stock + '.SA'].mean()*100))

print("\nStandard deviation of returns")
for stock in stocks:
    print('{} - {}'.format(stock, df_check[stock + '.SA'].std()*100))
```

```python
### Checking the dividend yield from each stock
```

```python
tickers.tickers['RENT3.SA'].dividends
```

```python
df_dividends = pd.DataFrame()

for ticker in tickers.tickers.keys():
    df_dividends_aux = tickers.tickers[ticker].dividends.to_frame().rename(columns={"Dividends": ticker})
    df_dividends = df_dividends.join(df_dividends_aux, how='outer')
```

```python
df_dividends
```

```python
df_dividends['year'] = df_dividends.index.year
df_dividends
```

```python
df_data['year'] = df_data.index.year
df_data
```

```python
df_dividends_year = df_dividends.groupby("year").sum()
df_dividends_year
```

```python
df_last_price = df_data.dropna().groupby("year").last()
df_last_price
```

```python
df_yield = df_dividends_year/df_last_price * 100
```

```python
df_yield.dropna(inplace=True)
df_yield
```

```python
## Average of each stock
df_yield.mean()
```

```python
## OVerall average
df_yield.mean().mean()
```

```python
df_return_year = df_last_price / df_last_price.shift(1) - 1
```

```python
## Checking as a global scope of stocks - Brazilian stock market
((df_return_year / 10).sum(axis=1)+1).cumprod().plot()
```

```python
strategy_return = []

for i in range(len(df_yield) - 1):
    wallet_stocks = df_yield.iloc[i, :].nlargest(5).index
    strategy_return += [(df_return_year[wallet_stocks].iloc[i + 1]/3).sum() +1]
```

```python
df_return_year[wallet_stocks]
```

```python
df_global = df_return_year / 10

fig, ax = plt.subplots()
pd.DataFrame(strategy_return, index=df_global.index[1:]).cumprod().plot(color="red",ax=ax)
(df_global.sum(axis=1)+1).cumprod().plot(color="blue",ax=ax)
```

A quick analysis through the graph I can visualize that the global wallet performed way better than the dividends wallet.  


However, when it came to a negative variation on the Brazilian market, the dividends wallet held a constant increasing value
once compared to the negative varation of the global wallet.
