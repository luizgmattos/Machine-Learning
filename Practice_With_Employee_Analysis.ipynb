# Answering Questions

# 1. Questions
- Total sum of employee's wage ( basic operation => summing 5 columns)

At this part the only work needed was to sum each column to find it's total and than combine the 3 values. 
The columns were:
 * Salary
 * Benefits
 * Taxes
 * VT
 * VR

```python
salario_total = funcionarios_df['Salario Base'].sum()
beneficios_total = funcionarios_df['Beneficios'].sum()
imposto_total = funcionarios_df['Impostos'].sum()
vt_total = funcionarios_df['VT'].sum()
vr_total = funcionarios_df['VR'].sum()
```

```python
print('Salario Total:',salario_total)
print('Beneficio Total:',beneficios_total)
print('Imposto Total:',imposto_total)
print('VT Total:',vt_total)
print('VR Total:',vr_total)
```

```python
total_gasto = salario_total+beneficios_total+imposto_total
print('Total money used with salary: R${:,}'.format(total_gasto))
```

Other way of making it: 

```python
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios']
+ funcionarios_df['VR'] + funcionarios_df['VT']
print('Total money used with salary: R${:,}'.format(funcionarios_df["Salario Total"].sum()))
```

# 2. Question

- What's the company income

To solve this problem, the csv files will be merged. In this case the client tables will be merged into the services tables.

The reason for this is, there can be only 1 client per ID where the same client could have multiple services, therefore the decision of merging into the service is reasonable

```python
faturamento_df = servicos_df.merge(clientes_df, on='ID Cliente')
faturamento_df
```

Next step will be to identify the total sum of each client. This will be done multiplying the monthly value column with the amount of time (in months) of the contract. 

All information will be stored in a new column:
 - Total Contrato


```python
faturamento_df['Total Contrato'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
faturamento_df
```

Next step it's to get the total sum of the value of contracts

```python
print('Total income: R${:,}'.format(faturamento_df["Total Contrato"].sum()))
```

# 3. Question

- What's the % of employees that closed a deal

To find this answer had to identify how many employees there were and the amount of employee that closed a contract.

For the amount of contract closed needed to avoid the duplicate values on the list. 

```python
contratos_fechados = len(servicos_df['ID Funcionário'].unique())
funcionarios_total = len(funcionarios_df['ID Funcionário'])
print('Amount of employees that closed a contract: {:.2%}'.format(contratos_fechados/funcionarios_total))
```

# 4. Question

- Calculating the areas that the company has closed deals

```python
area_contratos_df = servicos_df.merge(funcionarios_df, on='ID Funcionário')
area_contratos_df
```

```python
quantidade_contratos = area_contratos_df['Area'].value_counts()
quantidade_contratos
```

Visualizing the information as a graph

```python
import matplotlib.pyplot as plt

area_contratos_df["Area"].value_counts().plot(kind="bar")
plt.title("Contract area Frequency")
plt.xlabel("Area of Contract")
plt.ylabel("Frequency")
plt.xticks(rotation=0);
```
