### Checking sucessfull sales based on the profit median found earlier

In the next steps, I created the "sales_success" column following the Median of the "profit" variable found earlier when studying the amount sold * price of item.

Successful sales will only be consider on products that sold more than 7000 of it, which is according to the median.

```python
df_wish["sales_success"] = 0
df_wish.loc[df_wish["profit"] > 7000, "sales_success"] = 1
```

```python
df_wish["sales_success"].value_counts()
```

As seen through the value_counts, the sales_success column is almost 50% each value, meaning there is no need to balance this column using SMOTE

### Applying Machine Learning

```python
from sklearn.model_selection import train_test_split
```

```python
columns_eval = ['price','retail_price','uses_ad_boosts','rating',
                'badges_count', 'badge_product_quality', 'badge_fast_shipping',
                'product_variation_inventory','shipping_is_express','inventory_total',
                'has_urgency_banner','merchant_rating','variation','tags_count']
```

```python
x = df_wish[columns_eval]
y = df_wish["sales_success"]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
grid_param = [
    {'n_estimators':[3,10,30],
     'max_features': [2,4,6,8],
    },
    {'bootstrap':[False],
     'n_estimators':[3,10],
     'max_features': [2,3,4]
    }]
```

```python
rf_csf = RandomForestClassifier()
grid_search = GridSearchCV(rf_csf, grid_param, cv=5,
                          return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
rf_model = grid_search.best_estimator_
```

```python
x_test.iloc[0].to_frame().transpose()
```

```python
y_test.iloc[0]
```

```python
y_test.iloc[1]
```

```python
y_test.iloc[7]
```

```python
rf_model.predict(x_test.iloc[0].to_frame().transpose())
```

```python
rf_model.predict(x_test.iloc[7].to_frame().transpose())
```

```python
rf_model.predict(x_test.iloc[1].to_frame().transpose())
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = rf_model.predict(x_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

#### Checking the feature importance

```python
feature_importances = pd.DataFrame(rf_model.feature_importances_,
                                  index = x.columns,
                                  columns=['importance']).sort_values('importance', ascending=True)
fig, ax = plt.subplots(figsize=(20,8))
feature_importances.plot(kind="barh",ax=ax)
```

```python
import shap
```

```python
explainer = shap.TreeExplainer(rf_model)
shap_values = explainer.shap_values(x)
shap.summary_plot(shap_values[1],x)
```

Analysing shap:

* The X axis shows the impact that each feature has on the model.
* After the line in the middle (0.0) shows when the feature impacts in a positive way the model
* Before the line in the middle (0.0) shows when the feature impacts in a negative way the model
